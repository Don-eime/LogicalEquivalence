```python
def unique_edges_from_edge(e: Edge):
    """ Law names are same for forward and reverse laws"""
    n1_laws = [ej.law for ej in e.n1.get_edges]
    n2_laws = [ej.law for ej in e.n2.get_edges]

    n1_unique_laws = [law for law in n1_laws if law.short_name not in [law.short_name for law in n2_laws]]
    n2_unique_laws = [law for law in n2_laws if law.short_name not in [law.short_name for law in n1_laws]]

    return n1_unique_laws, n2_unique_laws
```

```python
def zip_edge_with_eligibilities_of_n1_and_n2(edge_list):
    return list(zip(edge_list, map(unique_edges_from_edge, edge_list)))
```

```python
# [(e, un[0], un[1]) for e, un in unique_edge_pairs if un[0] or un[1]]
```

```python
def edges_that_provide_new_options(edge_list):
    edge_elig_zip = zip_edge_with_eligibilities_of_n1_and_n2(edge_list)
    only_edges_with_unique_laws_on_atleast_one_side = [(e, un[0], un[1]) for e, un in edge_elig_zip if un[0] or un[1]]
    return only_edges_with_unique_laws_on_atleast_one_side
```

```python
def all_edges_from(node_list):
    edges = []
    for node in node_list:
        edges.extend(node.get_edges)
    return edges
```
